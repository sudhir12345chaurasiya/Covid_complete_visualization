```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
%matplotlib inline

import plotly
import plotly.express as px
import plotly.graph_objects as go
plt.rcParams['figure.figsize']=17,8     # this will increase the size of the graphs

import cufflinks as cf
import plotly.offline as pyo
from plotly.offline import init_notebook_mode,plot,iplot

import folium   #folium is used for the world map figure
```

```python
df = pd.read_excel('Covid cases in India.xlsx')
```

```python
df
```

```python
df.drop(['S. No.'],axis=1,inplace=True)
```

```python
df
```

```python
df['Total Cases']=df['Total Confirmed cases (Indian National)']+df['Total Confirmed cases ( Foreign National )']
```

```python
df
```

```python
total_cases_overall=df['Total Cases'].sum()
print('The total number of cases till now in India is ',total_cases_overall)
```

```python
df['Active Cases']=df['Total Cases']-(df['Death']+df['Cured'])
```

```python
df
```

### This way we can easily notice the highest values in the tables. 

```python
df.style.background_gradient(cmap='Reds')
```

## To list all the states with the number of total active cases in descending order

```python
Total_Active_Cases=df.groupby('Name of State / UT')['Total Cases'].sum().sort_values(ascending=False).to_frame()
```

# World Corona Cases

```python
world  = pd.read_csv('covid_19_data.csv')
world.head()
```

```python
#world.rename(columns={'Country/Region':'Country'},inplace=True)
#world.head()
```

```python
worldnew = world.filter(['Date','Province/State','Country/Region','Confirmed','Deaths','Recovered'], axis=1)
```

```python
#import numpy as np
#worldnew.replace(np.nan, 0, inplace=True)
worldnew.rename(columns={'Country/Region':'Country'},inplace=True)
worldnew.head()
```

```python
worldnew.query('Country=="India"')
```

```python
worldnew.groupby('Date').sum()
```

```python
confirmed=worldnew.groupby('Date').sum()['Confirmed'].reset_index()
death=worldnew.groupby('Date').sum()['Deaths'].reset_index()
rec=worldnew.groupby('Date').sum()['Recovered'].reset_index()
```

```python

fig=go.Figure()
fig.add_trace(go.Scatter(x=confirmed['Date'],y=confirmed['Confirmed'],mode='lines+markers',name='Confirmed',line=dict(color='blue',width=2)))

fig.add_trace(go.Scatter(x=death['Date'],y=death['Deaths'],mode='lines+markers',name='Deaths',line=dict(color='red',width=2)))
fig.add_trace(go.Scatter(x=rec['Date'],y=rec['Recovered'],mode='lines+markers',name='Recovered',line=dict(color='green',width=2)))
```
